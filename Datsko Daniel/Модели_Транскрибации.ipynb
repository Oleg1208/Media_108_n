```python
from vosk import Model, KaldiRecognizer, SetLogLevel
from pydub import AudioSegment
from pydub.utils import make_chunks
import wave
import pandas as pd
import time
import numpy
import subprocess
import json
import os
import zipfile
```

##Модель Vosk-ru-0.42 с аудио в mp3.

```python
!unzip -qo '/content/drive/MyDrive/vosk-model-ru-0.42.zip'
```

```python
def make_chunks(audio_segment, chunk_length_ms):
  chunks = []
  for i in range(0, len(audio_segment), chunk_length_ms):
    chunks.append(audio_segment[i:i + chunk_length_ms])
  return chunks
start_time = time.time()
Frame_rate = 16000
Channel = 1
model = Model('/content/vosk-model-ru-0.42')
rec = KaldiRecognizer(model, Frame_rate)
mp3 = AudioSegment.from_mp3('/content/drive/MyDrive/2023-10-02_14-22-14.172118_from_79842157659_to_00288_session_3258480198_talk.mp3')
mp3 = mp3.set_channels(Channel)
mp3 = mp3.set_frame_rate(Frame_rate)
chunk_size = 30 * 1000
chunks = make_chunks(mp3, chunk_size)
full_text = ""
for i, chunk in enumerate(chunks):
  rec.AcceptWaveform(chunk.raw_data)
  result = rec.Result()
  text = json.loads(result)["text"]
  full_text += text + " "
final_result = rec.FinalResult()
final_text = json.loads(final_result)["text"]
full_text += final_text
end_time = time.time()
elapsed_time = end_time - start_time
minutes = int(elapsed_time // 60)
seconds = int(elapsed_time % 60)
print(f"Затраченное время: {minutes} минут {seconds} секунд")
```

```python
from vosk import Model, KaldiRecognizer, SetLogLevel
from pydub.effects import low_pass_filter, high_pass_filter
from pydub.playback import play
from pydub import AudioSegment
from pydub.utils import make_chunks
from scipy.io import wavfile
from pydub.silence import split_on_silence
import noisereduce as nr
import numpy as np
import pyroomacoustics as pra
import wave
import pandas as pd
import time
import subprocess
import json
import os
import zipfile
```

```python
!unzip -qo '/content/drive/MyDrive/vosk-model-ru-0.10.zip'
```

##Нормализация эквализацией.

```python
def make_chunks(audio_segment, chunk_length_ms):
  chunks = []
  for i in range(0, len(audio_segment), chunk_length_ms):
    chunks.append(audio_segment[i:i + chunk_length_ms])
  return chunks
def high_pass_filter(audio_segment, low_freq):
  return audio_segment.high_pass_filter(low_freq)
def low_pass_filter(audio_segment, high_freq):
  return audio_segment.low_pass_filter(high_freq)
start_time = time.time()
Frame_rate = 16000
Channel = 1
low_freq = 300
high_freq = 3400
model = Model('/content/vosk-model-ru-0.10')
rec = KaldiRecognizer(model, Frame_rate)
mp3 = AudioSegment.from_mp3('/content/drive/MyDrive/2023-10-02_14-22-14.172118_from_79842157659_to_00288_session_3258480198_talk.mp3')
mp3 = mp3.set_channels(Channel)
mp3 = mp3.set_frame_rate(Frame_rate)
mp3 = high_pass_filter(mp3, low_freq)
mp3 = low_pass_filter(mp3, high_freq)
chunk_size = 60 * 1000
chunks = make_chunks(mp3, chunk_size)
full_text = ""
for i, chunk in enumerate(chunks):
  rec.AcceptWaveform(chunk.raw_data)
  result = rec.Result()
  text = json.loads(result)["text"]
  full_text += text + " "
final_result = rec.FinalResult()
final_text = json.loads(final_result)["text"]
full_text += final_text
end_time = time.time()
elapsed_time = end_time - start_time
minutes = int(elapsed_time // 60)
seconds = int(elapsed_time % 60)
print(f"Затраченное время: {minutes} минут {seconds} секунд")
```

```python
def make_chunks(audio_segment, chunk_length_ms):
  chunks = []
  for i in range(0, len(audio_segment), chunk_length_ms):
    chunks.append(audio_segment[i:i + chunk_length_ms])
  return chunks
start_time = time.time()
Frame_rate = 16000
Channel = 1
threshold = -20.0
ratio = 4
model = Model('/content/vosk-model-ru-0.10')
rec = KaldiRecognizer(model, Frame_rate)
mp3 = AudioSegment.from_mp3('/content/drive/MyDrive/2023-10-02_14-22-14.172118_from_79842157659_to_00288_session_3258480198_talk.mp3')
mp3 = mp3.set_channels(Channel)
mp3 = mp3.set_frame_rate(Frame_rate)
mp3 = mp3.compress_dynamic_range(threshold=threshold, ratio=ratio, attack=5.0, release=50.0)
chunk_size = 60 * 1000
chunks = make_chunks(mp3, chunk_size)
full_text = ""
for i, chunk in enumerate(chunks):
  rec.AcceptWaveform(chunk.raw_data)
  result = rec.Result()
  text = json.loads(result)["text"]
  full_text += text + " "
final_result = rec.FinalResult()
final_text = json.loads(final_result)["text"]
full_text += final_text
end_time = time.time()
elapsed_time = end_time - start_time
minutes = int(elapsed_time // 60)
seconds = int(elapsed_time % 60)
print(f"Затраченное время: {minutes} минут {seconds} секунд")
```

##Нормализация громкости.

```python
def make_chunks(audio_segment, chunk_length_ms):
  chunks = []
  for i in range(0, len(audio_segment), chunk_length_ms):
    chunks.append(audio_segment[i:i + chunk_length_ms])
  return chunks
def match_target_amplitude(sound, target_dBFS):
  change_in_dBFS = target_dBFS - sound.dBFS
  return sound.apply_gain(change_in_dBFS)
start_time = time.time()
Frame_rate = 16000
Channel = 1
target_dBFS = -20.0
model = Model('/content/vosk-model-ru-0.10')
rec = KaldiRecognizer(model, Frame_rate)
mp3 = AudioSegment.from_mp3('/content/drive/MyDrive/2023-10-02_14-22-14.172118_from_79842157659_to_00288_session_3258480198_talk.mp3')
mp3 = mp3.set_channels(Channel)
mp3 = mp3.set_frame_rate(Frame_rate)
mp3 = match_target_amplitude(mp3, target_dBFS)
chunk_size = 60 * 1000
chunks = make_chunks(mp3, chunk_size)
full_text = ""
for i, chunk in enumerate(chunks):
  rec.AcceptWaveform(chunk.raw_data)
  result = rec.Result()
  text = json.loads(result)["text"]
  full_text += text + " "
final_result = rec.FinalResult()
final_text = json.loads(final_result)["text"]
full_text += final_text
end_time = time.time()
elapsed_time = end_time - start_time
minutes = int(elapsed_time // 60)
seconds = int(elapsed_time % 60)
print(f"Затраченное время: {minutes} минут {seconds} секунд")
```

##Нормализация фильтрацией шума.

```python
def make_chunks(audio_segment, chunk_length_ms):
  chunks = []
  for i in range(0, len(audio_segment), chunk_length_ms):
    chunks.append(audio_segment[i:i + chunk_length_ms])
  return chunks
start_time = time.time()
Frame_rate = 16000
Channel = 1
model = Model('/content/vosk-model-ru-0.10')
rec = KaldiRecognizer(model, Frame_rate)
mp3 = AudioSegment.from_mp3('/content/drive/MyDrive/2023-10-02_14-22-14.172118_from_79842157659_to_00288_session_3258480198_talk.mp3')
mp3 = mp3.set_channels(Channel)
mp3 = mp3.set_frame_rate(Frame_rate)
audio_data = np.array(mp3.get_array_of_samples())
reduced_noise_audio = nr.reduce_noise(y=audio_data, sr=Frame_rate)
mp3 = AudioSegment(reduced_noise_audio.tobytes(), frame_rate=Frame_rate, sample_width=mp3.sample_width, channels=Channel)
chunk_size = 60 * 1000
chunks = make_chunks(mp3, chunk_size)
full_text = ""
for i, chunk in enumerate(chunks):
  rec.AcceptWaveform(chunk.raw_data)
  result = rec.Result()
  text = json.loads(result)["text"]
  full_text += text + " "
final_result = rec.FinalResult()
final_text = json.loads(final_result)["text"]
full_text += final_text
end_time = time.time()
elapsed_time = end_time - start_time
minutes = int(elapsed_time // 60)
seconds = int(elapsed_time % 60)
print(f"Затраченное время: {minutes} минут {seconds} секунд")
```

##Нормализация сегментацией и удалением тишины.

```python
def make_chunks(audio_segment, chunk_length_ms):
  chunks = []
  for i in range(0, len(audio_segment), chunk_length_ms):
    chunks.append(audio_segment[i:i + chunk_length_ms])
  return chunks
start_time = time.time()
Frame_rate = 16000
Channel = 1
model = Model('/content/vosk-model-ru-0.10')
rec = KaldiRecognizer(model, Frame_rate)
mp3 = AudioSegment.from_mp3('/content/drive/MyDrive/2023-10-02_14-22-14.172118_from_79842157659_to_00288_session_3258480198_talk.mp3')
mp3 = mp3.set_channels(Channel)
mp3 = mp3.set_frame_rate(Frame_rate)
chunks = split_on_silence(mp3, min_silence_len=500, silence_thresh=mp3.dBFS-14)
combined = AudioSegment.silent(duration=0)
for chunk in chunks:
  combined += chunk + AudioSegment.silent(duration=200)
fixed_chunks = make_chunks(combined, chunk_length_ms=60 * 1000)
full_text = ""
for i, chunk in enumerate(fixed_chunks):
  rec.AcceptWaveform(chunk.raw_data)
  result = rec.Result()
  text = json.loads(result)["text"]
  full_text += text + " "
final_result = rec.FinalResult()
final_text = json.loads(final_result)["text"]
full_text += final_text
end_time = time.time()
elapsed_time = end_time - start_time
minutes = int(elapsed_time // 60)
seconds = int(elapsed_time % 60)
print(f"Затраченное время: {minutes} минут {seconds} секунд")
```
