# Критерии классификации

Критерии `уникально-целевого` обращения:
1. Длительность звонка должна быть не менее `75 секунд`;
2. Телефонный номер абонента должен быть `уникальным`, т. е. его не должно быть в `CRM` заказчика. Либо, по нему не должно быть активности за последние `90 дней`. Исключение составляют `топовые площадки`, работающие по своим критериям (`ЦИАН`, `Яндекс.Недвижимость`, `Авито`, и т. д.);
3. Клиент должен знать минимальную информацию об объекте (понимать, куда он звонит): название `ЖК`, расположение, ценовую политику;
4. Номер абонента должен быть доступен в течение `15 дней` после совершённого звонка (при этом со стороны заказчика должно быть обеспечено `2 попытки` исходящего звонка в течение указанного срока);
5. Клиент должен быть `«адекватным»`. Не общаться на повышенных тонах, не употреблять ненормативную лексику и т. д.;
6. Клиент не должен быть повторным. Например, клиент `уже` купил квартиру и через `90 дней` решил купить машиноместо;
7. Заявка. В заявке должен быть указан номер телефона, а абонент при контакте с менеджером должен соответствовать всем критериям `уникально-целевого` обращения, перечисленным выше;
8. Озвученный общий бюджет покупки не должен быть `ниже 90%` от стоимости квартиры или коммерческого помещения, соответствующих площади/комнатности на момент обращения;
9. В случае, если клиент является уникальным, интересуется покупкой недвижимости, не является представителем партнёров или исполнителей и изъявил желание посетить офис продаж. При этом не обязательна фиксация уровня знания клиента об объекте в диалоге. При этом источник не должен относиться к каналам `«лидогенерация»` или `«тематические площадки»`;
10. Основной целью звонка клиента не должна являться покупка/аренда исключительно нежилого помещения, если это не являлось целью проводимой рекламной кампании.

Критерии НЕ `уникально-целевого` (вторичного) обращения:
1. Клиент позвонил повторно менее чем через `3 месяца` (`90 дней`) с момента последнего обращения;
2. Клиент, либо члены его семьи, уже купил (купили) `квартиру`/`машиноместо`/`кладовое помещение`, и хочет (хотят) совершить ещё одну покупку. В данном случае клиент относится к показателю `LTV` (показатель прибыли, которую компания получает от одного клиента за всё время работы с ним).

## Функции

```python
# Функция для сравнения столбцов в выбранных строках файла
def line_compare(filename, line_list, table=True):

  line_fields_list = []       # список со столбцами выбранных строк
  line_fields_count_list = [] # список с количеством столбцов выбранных строк
  nl = '\n'

  # Читаем файл
  with open(filename, 'r') as f:

    # Список строк
    lines = f.readlines()

    print(f"Файл {filename}:")
    print('-'*103)

    # Табличный вывод
    if table:

      print(end='   ')
      for i in range(len(line_list)):
        line_fields_list.append(lines[line_list[i]-1].split(';'))
        line_fields_count_list.append(len(line_fields_list[i]))
        print(f"{f'line {line_list[i]}: {line_fields_count_list[i]} fields':30s}", end='\t')
      print()
      print('-'*103)

      for j in range(max(line_fields_count_list)):
        print(f"{j+1:2d}", end=' ')
        for i in range(len(line_list)):
          if j < line_fields_count_list[i]:
            print(f"{f'{line_fields_list[i][j][:30] if len(line_fields_list[i][j]) > 30 else line_fields_list[i][j]}':30s}", end='\t')
          else:
            print(f"{'*':30s}", end='\t')
        print()

    # Построчный вывод
    else:

      for i in range(len(line_list)):
        line_fields_list.append(lines[line_list[i]-1].split(';'))
        line_fields_count_list.append(len(line_fields_list[i]))

      for j in range(max(line_fields_count_list)):
        print(f"{j+1}")
        for i in range(len(line_list)):
          if j < line_fields_count_list[i]:
            print(f"{line_fields_list[i][j].rstrip(nl)}")
          else:
            print(f"{'*'}")
        print('-'*103)

    print('-'*103)
    print('Количество строк в файле:', len(lines))
    print()

    return [lines[line_list[i]-1].rstrip(nl) for i in range(len(line_list))]
```

```python
# Функция для исправления проблемных строк и добавления их в датафрейм
def fix_bad_lines(df, filename, bad_lines):

  # Читаем файл
  with open(filename, 'r') as f:
    lines = f.readlines()

  # Названия столбцов
  keys = lines[4].rstrip('\n').split(';')

  # Проходим по списку с номерами bad_lines
  for i in bad_lines:

    # Создаём словарь для исправленной строки
    fix_line = dict()

    # Содержимое bad_line
    bad_line_values = lines[i-1].rstrip('\n').split(';')

    # Проходим по столбцам
    for j in range(len(keys)):

      if os.path.basename(filename) == 'Headliner.csv':

        if j < 61:
          fix_line[keys[j]] = bad_line_values[j]
        elif j == 61:
          fix_line[keys[j]] = bad_line_values[j]   + ';' + \
                              bad_line_values[j+1] + ';' + \
                              bad_line_values[j+2] + ';' + \
                              bad_line_values[j+3] + ';' + \
                              bad_line_values[j+4]
        else:
          fix_line[keys[j]] = bad_line_values[j+4]

      elif os.path.basename(filename) == 'Primavera .csv':

        if i in [800, 1347, 3373]:

          if j < 43:
            fix_line[keys[j]] = bad_line_values[j]
          elif j == 43:
            fix_line[keys[j]] = bad_line_values[j]   + ';' + \
                                bad_line_values[j+1] + ';' + \
                                bad_line_values[j+2] + ';' + \
                                bad_line_values[j+3] + ';' + \
                                bad_line_values[j+4]
          else:
            fix_line[keys[j]] = bad_line_values[j+4]

        elif i in [1050, 1289]:

          if j < 43:
            fix_line[keys[j]] = bad_line_values[j]
          elif j == 43:
            fix_line[keys[j]] = bad_line_values[j]   + ';' + \
                                bad_line_values[j+1] + ';' + \
                                bad_line_values[j+2] + ';' + \
                                bad_line_values[j+3] + ';' + \
                                bad_line_values[j+4] + ';' + \
                                bad_line_values[j+5] + ';' + \
                                bad_line_values[j+6] + ';' + \
                                bad_line_values[j+7] + ';' + \
                                bad_line_values[j+8] + ';' + \
                                bad_line_values[j+9]
          else:
            fix_line[keys[j]] = bad_line_values[j+9]

        else:

          if j < 43:
            fix_line[keys[j]] = bad_line_values[j]
          elif j == 43:
            fix_line[keys[j]] = bad_line_values[j]   + ';' + \
                                bad_line_values[j+1] + ';' + \
                                bad_line_values[j+2] + ';' + \
                                bad_line_values[j+3] + ';' + \
                                bad_line_values[j+4] + ';' + \
                                bad_line_values[j+5] + ';' + \
                                bad_line_values[j+6] + ';' + \
                                bad_line_values[j+7] + ';' + \
                                bad_line_values[j+8]
          else:
            fix_line[keys[j]] = bad_line_values[j+8]


      elif os.path.basename(filename) == 'Павелецкая_ЖК_«Павелецкая_сити».csv':

        if i in [18, 160, 673, 3403]:

          if j < 36:
            fix_line[keys[j]] = bad_line_values[j]
          elif j == 36:
            fix_line[keys[j]] = bad_line_values[j]   + ';' + \
                                bad_line_values[j+1] + ';' + \
                                bad_line_values[j+2]
          else:
            fix_line[keys[j]] = bad_line_values[j+2]

        elif i in [432, 739, 5181, 5183]:

          if j < 36:
            fix_line[keys[j]] = bad_line_values[j]
          elif j == 36:
            fix_line[keys[j]] = bad_line_values[j]   + ';' + \
                                bad_line_values[j+1] + ';' + \
                                bad_line_values[j+2] + ';' + \
                                bad_line_values[j+3] + ';' + \
                                bad_line_values[j+4]
          else:
            fix_line[keys[j]] = bad_line_values[j+4]

        else:

          if j < 36:
            fix_line[keys[j]] = bad_line_values[j]
          elif j == 36:
            fix_line[keys[j]] = bad_line_values[j]   + ';' + \
                                bad_line_values[j+1] + ';' + \
                                bad_line_values[j+2] + ';' + \
                                bad_line_values[j+3] + ';' + \
                                bad_line_values[j+4] + ';' + \
                                bad_line_values[j+5] + ';' + \
                                bad_line_values[j+6] + ';' + \
                                bad_line_values[j+7] + ';' + \
                                bad_line_values[j+8]
          else:
            fix_line[keys[j]] = bad_line_values[j+8]

    # Добавляем исправленную строку в датафрейм
    df = pd.concat([df, pd.Series(fix_line).to_frame().T], ignore_index=True)

  return df
```
