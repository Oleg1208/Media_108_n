# Экспорт таблицы

```python
# экспорт таблицы
!pip install gdown

import gdown
import pandas as pd

url = 'https://drive.google.com/uc?id=1om_ZmC0GzLAqmKv0Q2mhs3i-g03ZGn_8'
output = 'Headliner.csv'  # Укажите желаемое имя файла

gdown.download(url, output, quiet=False)

df = pd.read_csv(output, delimiter=';', on_bad_lines='warn', header = 5)

# Теперь 'df' содержит вашу таблицу
```

```python
df.isna().sum()  # количество пропусков в каждом столбце
```

```python
df.isna().sum() / df.shape[0] # доля пропусков от общего количества элементов:
```

```python
# общее количество пропусков во всей таблице
df.isna().sum().sum()
```

```python
# общее количество пропусков во всей таблице в процентах
proc = df.isna().sum().sum() / df.size
print('{:.1%}'.format(proc))
```
