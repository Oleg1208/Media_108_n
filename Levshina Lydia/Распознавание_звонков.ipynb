# Павелецкая (pavcity.turbo.site)

```python
gdown.download('https://drive.google.com/uc?id=1MT_e3qWuPEkUKCZVz6Z6Qnxuxeo2F-m9')
```

```python
path = 'Павелецкая (pavcity.turbo.site).csv'
data = pd.read_csv(path, sep=';', header = 4)
```

```python
data
```

```python
# форма датафрейма
data.shape
```

```python
# размер датафрейма
data.size
```

```python
# количество непустых записей в каждом столбце
data.count().head(80)
```

```python
# тип данных каждого столбца датафрейма
data.dtypes.head(20)
```

```python
# количество непустых записей, типы данных
data.info()
```

```python
# количество пропусков в каждом столбце датафрейма
data.isna().sum().head(73)
```

```python
data.isna().sum().head(30) / data.shape[0]
```

```python
# общее количество пропусков во всей таблице data
data.isna().sum().sum()
```

```python
# общее количество пропусков во всей таблице data
proc = data.isna().sum().sum() / data.size
print('Общее количество пропусков во всей таблице:', '{:.1%}'.format(proc))
```

## Удаляем из датафрейма пустые столбцы (метод dropna)

```python
data_ = data.dropna(axis = 1, how = 'all')
data_
```

```python
data_.info()
```

```python
print(data_.size)
print(data_.shape)
```

```python
# общее количество пропусков во всей таблице data_
data_.isna().sum().sum()
```

```python
# общее количество пропусков во всей таблице data_
proc_ = data_.isna().sum().sum() / data_.size
print('Общее количество пропусков во всей таблице:', '{:.1%}'.format(proc_))
```

```python
# описательные статистики
data_.describe()
```

```python
# описательные статистики для строковых данных
data_.describe(include=['O'])
```

```python
# отбор данных по столбцу "Длительность разговора"
data_['Длительность разговора']
```

```python
# отбор данных по столбцам "Длительность разговора", 'Теги'
data_[['Длительность разговора', 'Теги']]
```

```python
data_.Теги
```

```python
#data_.Длительность разговора
```

```python
# отбор строк с 1 до 4 (не включая)
data_[1:4]
```

```python
data_.loc[1:4, ['Длительность разговора', 'Теги']]
```

```python
# фильтр по столбцу 'Теги': ищем 'Нецелевой контакт'
data_[data_.Теги == 'Нецелевой контакт']
```

```python
# фильтр по столбцу 'Теги': ищем 'Целевой контакт'
data_[data_.Теги == 'Целевой контакт']
```

```python
# уникальные значения в столбце 'Теги'
data_.Теги.unique()
```

```python
# уникальные значения в столбце 'Теги'
data_.nunique()
```

```python
data_1 = data_.groupby(['Теги']).count()
display(data_1)
```
