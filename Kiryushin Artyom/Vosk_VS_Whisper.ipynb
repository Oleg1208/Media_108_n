```python
from pydub import AudioSegment

#Для прослушивания аудио-треков прямо на страницах collab ноутбуков
from IPython.display import Audio
```

```python
from vosk import Model, KaldiRecognizer, SetLogLevel
from pydub import AudioSegment
import subprocess
import json
import os

SetLogLevel(0)

# Проверяем наличие модели
if not os.path.exists("/content/drive/My Drive/Media_108/Vosk_модель/vosk-model-ru-0.42"):
    print ("Please download the model from https://alphacephei.com/vosk/models and unpack as 'model' in the current folder.")
    exit (1)

# Устанавливаем Frame Rate
FRAME_RATE = 16000
CHANNELS=1

model = Model("/content/drive/My Drive/Media_108/Vosk_модель/vosk-model-ru-0.42")
rec = KaldiRecognizer(model, FRAME_RATE)
rec.SetWords(True)

# Используя библиотеку pydub делаем предобработку аудио
mp3 = AudioSegment.from_mp3('/content/Запись 3098496016.mp3')
mp3 = mp3.set_channels(CHANNELS)
mp3 = mp3.set_frame_rate(FRAME_RATE)

# Преобразуем вывод в json
rec.AcceptWaveform(mp3.raw_data)
result = rec.Result()
```

```python
text_vosk = json.loads(result)["text"]
```

```python
Audio('/content/Запись 3098496016.mp3', rate=16000)
```

```python
# Добавляем пунктуацию

cased = subprocess.check_output('python3 recasepunc/recasepunc.py predict recasepunc/checkpoint', shell=True, text=True, input=text_vosk)

# Записываем результат в файл "data.txt"
with open('data.txt', 'w') as f:
    json.dump(cased, f, ensure_ascii=False, indent=4)
```

```python
import torch
import numpy as np
from transformers import BertTokenizer, BertModel
from scipy.spatial.distance import cosine

# Загрузка предобученной модели BERT и токенизатора
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
# Функция для преобразования текста в эмбеддинги с помощью BERT
def get_bert_embedding(text):
    input_ids = tokenizer.encode(text, add_special_tokens=True, return_tensors='pt')
    with torch.no_grad():
        output = model(input_ids)
        last_hidden_states = output.last_hidden_state
        mean_last_hidden_states = torch.mean(last_hidden_states, 1)
    return mean_last_hidden_states
```

```python
# Пример текстов для сравнения
text1 = "The sky is blue"
text2 = "The sun is shining in the sky"

# Получение эмбеддингов для каждого текста
embedding1 = get_bert_embedding(text1)
embedding2 = get_bert_embedding(text2)

# Вычисление косинусного расстояния между эмбеддингами
similarity = 1 - cosine(embedding1, embedding2)
print(f"Схожесть текстов: {similarity}")
```

## DistilBERT (Совпадение 98%)

```python
import torch
from transformers import DistilBertTokenizer, DistilBertModel
from scipy.spatial.distance import cosine
```

```python
# Загрузка предобученной модели DistilBERT и токенизатора
model_name = 'distilbert-base-uncased'
tokenizer = DistilBertTokenizer.from_pretrained(model_name)
model = DistilBertModel.from_pretrained(model_name)
```

```python
# Функция для получения эмбеддингов текстов с использованием модели DistilBERT
def get_distilbert_embeddings(texts):
    inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
        embeddings = torch.mean(outputs.last_hidden_state, 1)
    return embeddings
```

```python
# Получение эмбеддингов для каждого текста
embeddings = get_distilbert_embeddings([text1, text2])

# Вычисление косинусного расстояния между эмбеддингами
similarity = 1 - cosine(embeddings[0], embeddings[1])
print(f"Схожесть текстов: {similarity}")
```

## NLTK (Совпадение 98%)

Для вычисления косинусного сходства между двумя фрагментами текста

Этот код сначала токенизирует и лемматизирует тексты, удаляет стоп-слова , а затем создает векторы TF-IDF для текстов. Наконец, он вычисляет косинусное сходство между векторами, используя функцию cosine_similarity из sklearn.metrics.pairwise .

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
nltk.download('punkt')
```

```python
nltk.download('stopwords')
```

```python
def func_text_similarity(text1, text2):
    # Tokenize and lemmatize the texts
    tokens1 = word_tokenize(text1)
    tokens2 = word_tokenize(text2)
    lemmatizer = WordNetLemmatizer()
    tokens1 = [lemmatizer.lemmatize(token) for token in tokens1]
    tokens2 = [lemmatizer.lemmatize(token) for token in tokens2]

    # Remove stopwords
    stop_words = stopwords.words('english')
    tokens1 = [token for token in tokens1 if token not in stop_words]
    tokens2 = [token for token in tokens2 if token not in stop_words]

    # Create the TF-IDF vectors
    vectorizer = TfidfVectorizer()
    vector1 = vectorizer.fit_transform(tokens1)
    vector2 = vectorizer.transform(tokens2)

    # Calculate the cosine similarity
    similarity = cosine_similarity(vector1, vector2)

    return similarity
```

```python
text_similarity = func_text_similarity(text_whisper, text_vosk)
```

```python
# Нахождение среднего значения только для значений выше порога
threshold = 0.01 # Установка порога для схожести
above_threshold_similarity = text_similarity[text_similarity > threshold]
mean_above_threshold_similarity = np.mean(above_threshold_similarity)
print(f'Средняя схожесть выше порога {threshold}: {mean_above_threshold_similarity}')
```

```python
import torch
```

```python
def str2int(text):
    # a temporary list to store the string labels
    temp_list = text.tolist()

    # dictionary that maps integer to its string value
    label_dict = {}

    # list to store integer labels
    int_labels = []

    for i in range(len(temp_list)):
        label_dict[i] = temp_list[i]
        int_labels.append(i)

    return int_labels
```

```python
# Calculate the cosine similarity between two texts
def cosine_similarity(text1, text2):
    # Convert the texts to tensors

    text1 = torch.tensor([text1])
    text2 = torch.tensor([text2])

    # Calculate the dot product of the texts
    dot_product = torch.matmul(text1, text2.transpose(1, 0))

    # Calculate the norms of the texts
    norm1 = torch.norm(text1, dim=1)
    norm2 = torch.norm(text2, dim=1)

    # Calculate the cosine similarity
    cosine_similarity = dot_product / (norm1 * norm2)

    return cosine_similarity
```

```python
similarity1 = cosine_similarity(text1, text2)

print(f"Similarity between text1 and text2: {similarity1:.2f}")
```
